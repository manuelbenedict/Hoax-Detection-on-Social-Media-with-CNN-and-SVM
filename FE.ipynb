#### TF-IDF Vectorizing

```python
%%time
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

#define tf-idf
tf_idf_vec = TfidfVectorizer(ngram_range=(1,2), max_features=10000)

#transform
tf_idf_data = tf_idf_vec.fit_transform(df_tweetberita['Text'])

#create dataframe
tf_idf_df  = pd.DataFrame(tf_idf_data.toarray(),columns=tf_idf_vec.get_feature_names())
features_v = tf_idf_vec.get_feature_names()
vector     = tf_idf_df.values.tolist()
df_tweetberita['TFIDF_vec'] = vector
tf_idf_df
```

## Feature Expansion

<img src="../images/Feature Expansion.PNG" width="40%">

```python
def checkWords(i, j):
    try:
        # pengecekan if one word in W appears in tweet T
        a = [t for t in d[features_v[j]] if t in [i]]
        return a
    except:
        return []
    
def weightCheck(w, i, j):
    try:
        if w != []:
            return vector[i][features_v.index(w[0])]
        else:
            return vector[i][j]
    except:
        return vector[i][j]
```

```python
pd.set_option("display.max_columns", None)
```

```python
%%time
fe_vec = []
print()
print('============ FEATURE EXPANSION IS PROCESSING ============')
for i in range(len(vector)):
    v = [weightCheck(checkWords(i,j), i, j) if (vector[i][j] == 0.0) else vector[i][j] for j in range(len(vector[i]))]
    fe_vec.append(v)
df_tweetberita['TFIDF_vec_fe'] = fe_vec
fe_TFIDF_df = pd.DataFrame(fe_vec, columns=features_v)
fe_TFIDF_df
```
