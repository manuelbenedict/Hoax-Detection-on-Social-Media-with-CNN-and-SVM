```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Setting variables to be used below
maxTweets = 2000

# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('PC Tidak Ditahan').get_items()):
    if i>maxTweets:
        break
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.displayname, tweet.user.location,
                        tweet.user.followersCount, tweet.user.friendsCount, tweet.user.verified, tweet.retweetCount, tweet.lang,
                        tweet.media, tweet.user.description, tweet.user.profileImageUrl])
```

```python
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'DisplayName', 'Location', 'Followers',
                                                'Following', 'Verified', 'Retweet', 'Language', 'Media', 'Description', 'ProfileImage'])

# Display dataframe
tweets_df
```

```python
tweets_df.to_csv("dataset_PCTidakDitahan.csv")  
```
