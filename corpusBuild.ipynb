```python
df['TokenizedText'] = df['text']
df["TokenizedText"] = df.apply(lambda row: nltk.word_tokenize(row['TokenizedText']), axis=1)
lines = list(df['TokenizedText'])
df['TokenizedText'] = lines
df
```

```python
tweetberita = df['TokenizedText'].astype(str).apply(lambda x: ast.literal_eval(x))
```

```python
%%time
from glove import Corpus, Glove

#Creating a corpus object
corpus_tweetberita = Corpus()

#Training the corpus to generate the co occurence matrix which is used in GloVe
corpus_tweetberita.fit(tweetberita, window=10)

glove = Glove(no_components=100, learning_rate=0.05) 
glove.fit(corpus_tweetberita.matrix, epochs=30, no_threads=4, verbose=True)
glove.add_dictionary(corpus_tweetberita.dictionary)
glove.save('corpus_tweetberita.model')
```

```python
def similarity(feature,n,model):
    """
    INPUT  : kata yg ingin dicari similaritynya, jumlah top-n similaritynya, model GloVe yang ingin digunakan
    OUTPUT : list similarity dari input
    DESC   : function untuk mengembalikan list kata similarity dari suatu kata. jika corpus similarity tidak bisa mengenali kata
             tersebut akan mereturn list kosong.
    """
    if model == 'tweetberita':                                            # model yang digunakan menggunakan dataset tweet
        try:                                                              # try condition
            sim_word = glove.most_similar(feature, number=n)         # ambil similarity dari fitur
            word = [sim_word[i][0] for i in range(len(sim_word))]         # masukin ke list similaritynya
            return word                                                   # return list similarity word
        except:                                                           # except condition jikalau try condition mendapati error
            word = []                                                     # inisialisasi list kosong
            return word                                                   # return list kosong
```

```python
k = 0
for i in df_similarity['Words'] :
  glove.most_similar(df_similarity['Words'][k],number=16)
  sim = similarity(df_similarity['Words'][k],16,'tweetberita')
  df_similarity['Rank 1'][k] = sim[0]
  df_similarity['Rank 2'][k] = sim[1]
  df_similarity['Rank 3'][k] = sim[2]
  df_similarity['Rank 4'][k] = sim[3]
  df_similarity['Rank 5'][k] = sim[4]
  df_similarity['Rank 6'][k] = sim[5]
  df_similarity['Rank 7'][k] = sim[6]
  df_similarity['Rank 8'][k] = sim[7]
  df_similarity['Rank 9'][k] = sim[8]
  df_similarity['Rank 10'][k] = sim[9]
  df_similarity['Rank 11'][k] = sim[10]
  df_similarity['Rank 12'][k] = sim[11]
  df_similarity['Rank 13'][k] = sim[12]
  df_similarity['Rank 14'][k] = sim[13]
  df_similarity['Rank 15'][k] = sim[14]
  k = k+1
```
